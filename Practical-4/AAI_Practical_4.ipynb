```python
import pandas as pd

# Sample data: We assume we have 100 observations
data = {
    'Event_A': [1, 1, 1, 0, 1, 0, 1, 0, 0, 1] ,
    'Event_B': [1, 0, 1, 1, 0, 1, 1, 0, 1, 0]
}

df = pd.DataFrame(data)

df
```

```python
# Calculate joint probability
joint_count = df[(df['Event_A'] == 1) & (df['Event_B'] == 1)].shape[0]
total_count = df.shape[0]
joint_probability = joint_count / total_count

print(f'Joint Probability P(A and B): {joint_probability}')

```

```python
# Calculate conditional probability
# P(A | B) = P(A and B) / P(B)

# Probability of B
prob_B = df[df['Event_B'] == 1].shape[0] / total_count

# Conditional probability P(A | B)
conditional_probability = joint_probability / prob_B

print(f'Conditional Probability P(A | B): {conditional_probability}')
```
